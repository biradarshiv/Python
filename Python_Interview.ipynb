```python
# 21. Create a generator to produce first n prime numbers
def generate_primes(n):
    count = 0
    num = 2
    while count < n:
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            yield num
            count += 1
        num += 1

n = 50
prime_generator = generate_primes(n)
for prime in prime_generator:
    print(prime, end=' ')
```

```python
# Decorators are used to modify the behavior of functions or methods without changing their code.
# Decorator function is a function that takes another function as its argument.

def decor_result(result_function):
    def distinction(marks):
        results = [] # List to store results for each element
        for m in marks:
            if m >= 75:
                print("DISTINCTION")
            results.append(result_function([m])) # pass a list with the current mark
        return results # Return the collected results
    return distinction

@decor_result
def result (marks):
    for m in marks:
        if m >= 33:
            pass
        else:
            print ("FAIL")
            return "FAIL" # Return FAIL if any element fails
    else:
        print("PASS")
        return "PASS" # Return PASS if all elements pass

# Get the resuLis for each eLeinent
results = result([45, 78, 80, 34, 66, 90, 30])

print("Results:", results)

```

This code demonstrates a Decorator in Python. A decorator is essentially a "wrapper" around a function that allows you to execute code before and after the original function runs.
Here is a step-by-step breakdown of how this specific code works:
1. The Decorator (decor_result)
This function is designed to "intercept" the result function.
	•	The Argument (result_function): It takes the function being decorated (in this case, result) as an input.
	•	The Inner Function (distinction): This is the actual "wrapper." It replaces the original result function logic with its own.
	•	The Logic: It loops through the marks before the main function does. If a mark is $\ge 75$, it prints "DISTINCTION".
	•	Calling the Original: Inside the loop, it calls result_function([m]). This executes the original "Pass/Fail" logic for every single number in the list.
2. The Decorated Function (result)
The @decor_result symbol tells Python: "Whenever result() is called, send the arguments to decor_result first."
	•	Its primary job is to check if a mark is $\ge 33$ (PASS) or less (FAIL).
	•	Because of the decorator, this function is now being called multiple times (once for each mark in the list).
3. The Execution Flow
When you run result([45, 78, 80, ...]), the following happens:
	1	Entry: The list of marks enters the distinction function inside the decorator.
	2	Iteration 1 (Mark: 45): * Is $45 \ge 75$? No.
	◦	Calls result([45]). Since $45 > 33$, it prints PASS and returns "PASS".
	3	Iteration 2 (Mark: 78): * Is $78 \ge 75$? Yes! Prints DISTINCTION.
	◦	Calls result([78]). Prints PASS and returns "PASS".
	4	Iteration 7 (Mark: 30):
	◦	Is $30 \ge 75$? No.
	◦	Calls result([30]). Since $30 < 33$, it prints FAIL and returns "FAIL".
4. The Output
Based on your input list [45, 78, 80, 34, 66, 90, 30], the console will look like this:
Plaintext

PASS
DISTINCTION
PASS
DISTINCTION
PASS
PASS
PASS
DISTINCTION
PASS
FAIL
Results: ['PASS', 'PASS', 'PASS', 'PASS', 'PASS', 'PASS', 'FAIL']
Key Concept: "Modifying Behavior"
Without changing a single line of code inside the result function, you successfully added a "Distinction" check. This is the power of decorators: they allow you to keep your core logic (result) clean and simple, while adding extra features (distinction) externally.


## In the def result (marks): when will it enter the second else condition.

In the result function, the second else belongs to the for loop, not to an if statement. This is a special Python feature called the for-else block.
The code will enter that second else condition only if the for loop finishes naturally without being stopped by a return or break statement.
How it works in your code:
	1	The Loop starts: It checks every mark m in the list marks.
	2	The Trigger: If it finds a mark that is less than 33, it executes the else inside the loop:
	◦	It prints "FAIL".
	◦	It executes return "FAIL".
	◦	The function ends immediately. Because it returned, the loop was "broken," and the second else is skipped.
	3	The "Success" Path: If it goes through the entire list and every single mark is 33 or higher:
	◦	The for loop completes its task for every item.
	◦	Because the loop was never interrupted by a return, Python enters the for-else block.
	◦	It prints "PASS" and returns "PASS".
Simplified Summary:
	•	Enter else: If every student passed (Loop finished).
	•	Skip else: If at least one student failed (Loop was interrupted by return "FAIL").
A Common Source of Confusion
In most languages, else only follows if. In Python, when you see else at the same indentation level as for, think of it as "if the loop reached the end."

```python
# 33. What are generators in python? Create a generator for first n natural even numbers.

# Generators in Python are a type of iterable, like lists or tuples, but they allow you to iterate over
# their elements lazily, one at a time, on-the-fly. This means that generators generate values as you
# need them, rather than storing all values in memory at once. This can be highly memory-efficient,
# especially when dealing with large datasets or when generating an infinite sequence.

def evenNum(n):
    i = 1
    while n:
        yield 2*i
        i+=1
        n-=1

it = evenNum(10)
even_list=[]

while True:
    try:
        even_list.append (next (it) )
    except StopIteration:
        break
print(even_list)
```

```python
# 34. Is Function overloading allowed in python ?

# In Python, function overloading as seen in languages like C++ or Java is not directly supported.
# However, Python provides several ways to achieve similar behavior through default arguments,
# variable-length arguments, and more advanced techniques like using functools.singledispatch.

# Here are a few methods to achieve function overloading in Python:

print('1. Using Default Arguments')
#  You can use default arguments to provide different behavior based on the number of arguments passed.

def greet(name, greeting="Hello"):
    return f"{greeting}, {name}! "

print(greet("Ashutosh"))
print(greet("Bob", "Hi"))

print('\n2. Using Variable-Length Arguments')
# You can use *args and **kwargs to accept a variable number of arguments.

def add(*t):
    return sum(t)

print(add(1,3))
print(add(1,3,5,6))

print('\n3. Using functools.singledispatch')

# The functools.singledispatch decorator allows you to create a single-dispatch generic function,
# which can have different implementations based on the type of the first argument.
from functools import singledispatch

@singledispatch
def process (value):
    raise NotImplementedError ("Unsupported type")

@process.register(int)
def _(value):
    return f"Processing an integer: {value}"

@process.register(str)
def _(value):
    return f"Processing a string: {value}"

print(process(10))
print(process("hello"))

print('\n4. Using Class Methods')
#You can also achieve overloading behavior using class methods.
class Math:
    @staticmethod
    def multiply(a,b):
        return a*b
    @staticmethod
    def multiply(a,b,c):
        return a*b*c

#python does not support method over loading, so the Last defined method will be considered valid for only 3 arguments
#print (Math.multiply(2,3)) # this will raise an error
print (Math.multiply(2,3,4))

#To achieve overloading behavior you can use different method names or variable-length arguments
class Math:
    @staticmethod
    def multiply(*t):
        result = 1
        for num in t:
            result *= num
        return result

print (Math.multiply(2,3))
print (Math.multiply(2,3,4,5))

```

```python
# First 10 fibonacci numbers which are also prime numbers

import math

def is_prime(n):
    """Efficiently checks if a number is prime."""
    if n < 2: return False
    if n == 2 or n == 3: return True
    if n % 2 == 0 or n % 3 == 0: return False
    
    # Check factors from 5 up to sqrt(n)
    for i in range(5, int(math.sqrt(n)) + 1, 6):
        if n % i == 0 or n % (i + 2) == 0:
            return False
    return True

def get_fibonacci_primes(limit):
    """Generates the first 'limit' numbers that are both Fibonacci and Prime."""
    fib_primes = []
    a, b = 0, 1  # Starting terms of Fibonacci sequence
    
    while len(fib_primes) < limit:
        if is_prime(a):
            fib_primes.append(a)
        # Calculate next Fibonacci number
        a, b = b, a + b
    
    return fib_primes

# Execute and print the first 10
result = get_fibonacci_primes(10)
print(f"First 10 Fibonacci Primes: {result}")
```

```python
# 64. Check if a number is palindrome or not.
def is_palindrome(n):
    original = n
    reversed_num = 0
    while n > 0:
        reversed_num = reversed_num * 10 + n % 10
        n //= 10 # // (Floor Division): It divides the number by 10 and rounds down to the nearest whole integer (dropping the decimal part).
    return original == reversed_num

# Example usage:
number = 12321
if is_palindrome(number):
    print(f"{number} is a palindrome.")
else:
    print(f"{number} is not a palindrome.")
```

```python
# 65. Check if a number is spy number. Means sum of its digits equal to the product of its digits.
def is_spy_number(n):
    sum_digits = 0
    product_digits = 1
    while n > 0:
        digit = n % 10
        sum_digits += digit
        product_digits *= digit
        n //= 10
    return sum_digits == product_digits

# Example usage:
number = 123
if is_spy_number(number):
    print(f"{number} is a spy number.")
else:
    print(f"{number} is not a spy number.")
```

```python
# 81. What are the commonly used decorators in python ?
# Decorators are a powerful feature in Python that allow you to modify the behavior of functions or classes without changing their source code. Some commonly used decorators in Python include:

"""
1. @staticmethod: This decorator is used to define a static method in a class. Static methods do not require an instance of the class to be called and do not have access to the instance (self) or class (cls) variables.
2. @classmethod: This decorator is used to define a class method in a class. Class methods receive the class (cls) as the first argument instead of the instance (self). They can access and modify class state that applies across all instances of the class.
3. @property: This decorator is used to define a property in a class. It allows you to define methods that can be accessed like attributes. This is useful for creating read-only attributes or for adding logic when getting or setting an attribute.
4. @functools.wraps: This decorator is used to preserve the metadata of the original function when creating a decorator. It helps to maintain the original function's name, docstring, and other attributes.
5. @lru_cache: This decorator is used to cache the results of a function based on its arguments. It can improve performance by avoiding redundant calculations for previously seen inputs.
6. @contextmanager: This decorator is used to define a context manager in Python. It allows you to create a block of code that can be used with the 'with' statement, providing a way to manage resources such as file handling or database connections.
7. @dataclass: This decorator is used to automatically generate special methods like __init__, __repr__, and __eq__ for classes that primarily store data. It simplifies the creation of classes that are meant to be used as data containers.
8. @abstractmethod: This decorator is used in abstract base classes to indicate that a method must be implemented by any subclass. It is part of the abc module and helps to enforce a certain interface for subclasses.
9. @singledispatch: This decorator is used to create a single-dispatch generic function. It allows you to define a function that can have different behaviors based on the type of the first argument.
10. @total_ordering: This decorator is used to simplify the implementation of rich comparison methods in a class. By defining a few of the comparison methods, the decorator can automatically generate the rest of them.
"""

# provide examples for each of the above decorators.
# 1. @staticmethod
print("\nExample of @staticmethod:\n")
class MyClass:
    @staticmethod
    def static_method():
        return "This is a static method."
print(MyClass.static_method())  # Output: This is a static method.


# 2. @classmethod
print("\nExample of @classmethod:\n")
class MyClass:
    class_variable = "This is a class variable."
    @classmethod
    def class_method(cls):
        return f"This is a class method. {cls.class_variable}"
print(MyClass.class_method())  # Output: This is a class method. This is a class variable.


```

```python

# 3. @property
print("\nExample of @property:\n")
class MyClass:
    def __init__(self, value):
        self._value = value
    @property
    def value(self):
        return self._value
    @value.setter
    def value(self, new_value):
        self._value = new_value
obj = MyClass(10)
print(obj.value)  # Output: 10
obj.value = 20
print(obj.value)  # Output: 20


# 4. @functools.wraps
print("\nExample of @functools.wraps:\n")
import functools
def my_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print("Before the function call.")
        result = func(*args, **kwargs)
        print("After the function call.")
        return result
    return wrapper
@my_decorator
def say_hello(name):
    """This function says hello."""
    return f"Hello, {name}!"
print(say_hello("Alice"))  # Output: Before the function call. Hello, Alice! After the function call.
print(say_hello.__name__)  # Output: say_hello
print(say_hello.__doc__)   # Output: This function says hello.
```

```python
# 5. @lru_cache
print("\nExample of @lru_cache:\n") 
import functools
@functools.lru_cache(maxsize=None)
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(10))  # Output: 55

# 6. @contextmanager
print("\nExample of @contextmanager:\n")
from contextlib import contextmanager
@contextmanager
def open_file(file_name, mode='r'):
    file = open(file_name, mode)
    try:
        yield file
    finally:
        file.close()
with open_file('example.txt', 'w') as f:
    f.write("Hello, World!")
```

```python
# 82.different types of parameters in python.
# In Python, there are several types of parameters that can be used in function definitions. These include:
# 1. Positional Parameters: These are the most common type of parameters. They are defined in the function signature and must be passed in the correct order when calling the function.
def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")
greet("Alice", 30)  # Output: Hello, Alice! You are 30 years old.
# 2. Default Parameters: These parameters have default values assigned to them. If a value is not provided for a default parameter when calling the function, the default value will be used. Default parameters must be defined after all positional parameters.
def greet(name, age=25):
    print(f"Hello, {name}! You are {age} years old.")
greet("Bob")  # Output: Hello, Bob! You are 25 years old.
greet("Charlie", 35)  # Output: Hello, Charlie! You are 35 years old.
# 3. Keyword Parameters: These parameters are passed by explicitly specifying the parameter name and value when calling the function. This allows you to pass arguments in any order.
def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")
greet(age=40, name="David")  # Output: Hello, David! You are 40 years old.
# 4. Variable-length Parameters: These parameters allow you to pass a variable number of arguments to a function. There are two types of variable-length parameters:
#    a. *args: This allows you to pass a variable number of positional arguments to a function. The arguments are collected into a tuple.
def greet(*args):
    for name in args:
        print(f"Hello, {name}!")
greet("Alice", "Bob", "Charlie")  # Output: Hello, Alice!
                                     #         Hello, Bob!
                                     #         Hello, Charlie!
#    b. **kwargs: This allows you to pass a variable number of keyword arguments to a function. The arguments are collected into a dictionary.
def greet(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
greet(name="Alice", age=30, city="New York")  # Output: name: Alice
                                               #         age: 30
                                               #         city: New York
# 5. Positional-only Parameters: These parameters can only be passed positionally and cannot be used as keyword arguments. They are defined using a / in the function signature.
def greet(name, age, /):
    print(f"Hello, {name}! You are {age} years old.")
greet("Alice", 30)  # Output: Hello, Alice! You are 30 years old.
# greet(name="Bob", age=25)  # This will raise a TypeError because name and age are positional-only parameters.
# 6. Keyword-only Parameters: These parameters can only be passed as keyword arguments and cannot be used positionally. They are defined using a * in the function signature.
def greet(*, name, age):
    print(f"Hello, {name}! You are {age} years old.")
greet(name="Alice", age=30)  # Output: Hello, Alice! You are 30 years old.
# greet("Bob", 25)  # This will raise a TypeError because name and age are keyword-only parameters.

```

```python
# 83.Operator overloading in python.

# Operator overloading in Python allows you to define custom behavior for built-in operators (like +, -, *, etc.) when they are used with instances of your classes. 
# This is achieved by implementing special methods in your class that correspond to the operators you want to overload. 
# For example, to overload the addition operator (+), you would implement the __add__ method in your class.
# Here's an example of operator overloading in Python:
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        return NotImplemented

    def __str__(self):
        return f"Vector({self.x}, {self.y})"
v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2
print(v3)  # Output: Vector(4, 6)
# In this example, we have defined a Vector class with an __add__ method that allows us to add two Vector instances together using the + operator. The __str__ method is also implemented to provide a string representation of the Vector instance when printed. 
```

```python

# The __add__ method allows you to use the + operator between two Point instances.
# The __mul__ method allows you to use the * operator for both Point instances and scalar multiplication. 
# The __rmul__ method allows for right-hand scalar multiplication. 
# The __rmul__ method is called when the left operand does not support the multiplication operation with the right operand, and the right operand is of a type that supports the operation.
# The __repr__ method provides a string representation of the Point object.


class Point:
    def __init__ (self,x=0,y=0):
        self.x = x
        self.y = y

    def __add__(self, other):
            return Point(self.x + other.x, self.y + other.y)

    def __mul__(self, other):
        if isinstance(other, Point): # isinstance used to check if object is instance of a class or not
            return Point(self.x * other.x, self.y * other.y) 
        else:
            return Point(self.x * other, self.y * other)
    def __rmul__(self, other):
        return Point(self.x * other, self.y * other)
    def __repr__(self):
        return ("{0},{1}".format(self.x,self.y))
    
p1 = Point(2,3)
p2 = Point(3, 4)

print(p1 + p2)
print(p1 * p2)
```

```python
# 84. When you encounter a line that starts with "X-DSPAM-Confidence:" pull apart the line to
# extract the floating-point number on the line. Count these lines and then compute the total of
# the spam confidence values from these lines. When you reach the end of the file, print out the
# average spam confidence.

fname = 'mbox-short.txt' # input("Enter the file name: ")
try:
    fhand = open(fname)
except:
    print("File cannot be opened:", fname)
    exit()
count = 0
total_confidence = 0.0
for line in fhand:
    line = line.rstrip()
    if line.startswith("X-DSPAM-Confidence:"):
        count += 1
        confidence_value = float(line.split(":")[1].strip())
        total_confidence += confidence_value
if count > 0:
    average_confidence = total_confidence / count
    print(f"Average spam confidence: {average_confidence}")
else:
    print("No lines with 'X-DSPAM-Confidence:' found.")
```
