```python
# 21. Create a generator to produce first n prime numbers
def generate_primes(n):
    count = 0
    num = 2
    while count < n:
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            yield num
            count += 1
        num += 1

n = 50
prime_generator = generate_primes(n)
for prime in prime_generator:
    print(prime, end=' ')
```

```python
#program to check if a given number is prime or not
def prime_no(n):
    Flag = False
    if n < 2:
        return Flag
    else:
        for i in range(2,n):
            if n%i==0:
                return Flag
            else:
                Flag = True
    return Flag

num = int(input("Enter a number to whether it is prime or not: "))
prime_no(num)

#Enter a number: 37 True 
#Enter a number: 25 False 
```

```python
# 28. Global and local variable

x = 5 # global variable

def f1():
    global x
    x = 15 # global variable updated
    y = 10 # Local variable
    print("x=%d y=%d"%(x,y))

f1()
print(x)


x = 5 # global variable
def f2():
    #global x
    x = 15 # global variable updated
    y = 10 # Local variable
    print("x=%d y=%d"%(x,y))

f2()
print(x)
```
